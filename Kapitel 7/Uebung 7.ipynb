```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

## Aufgabe 1

```python
x = []
y = []
for i in range(1000):
    x.append(random.randint(-100,100))
    y.append(random.randint(-100,100))
```

```python
plt.plot(x, y, "b.")
plt.axis("equal")
#plt.axis([-100,100,-100,100])
plt.show()
```

### Versuch Punkte zufällig einzufärben (hat nicht funktioniert)

```python
farben = ['b', 'g', 'r', 'c', 'm', 'y']

x = []
y = []
for i in range(1000):
    x.append(random.randint(-100,100))
    y.append(random.randint(-100,100))
    plt.plot(x, y, f"{np.random.choice(farben)}.")
    
plt.axis("equal")
plt.axis([-100,100,-100,100])
plt.show()
```

## Aufgabe 2

```python
def f(x, y):
    return np.exp(-x**2) * np.sin(y)


x = np.linspace(-20, 20, 101)

y = np.linspace(-20, 20, 101)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)

plt.pcolormesh(X, Y, Z)

plt.show()
```
